```julia
using Random, Dates, Optimization, ComponentArrays, Lux, OptimizationOptimisers, DiffEqFlux,
      OrdinaryDiffEq, CSV, DataFrames, Dates, Statistics, Plots, DataDeps
```

```julia
data=CSV.read("/Users/seansantellanes/Downloads/EUG.csv",DataFrame)
data=data[:,Not(end)]
```

```julia
Features=[:"Temperature",:"Dew Point",:"Relative Humidity",:"Wind Speed"]
Units=["Faren","Faren","%","mph"]
Feature_names=["Temperature","Dew Point","Relative Humidity","Wind Speed"]
function plot_data(df)
    plots=map(enumerate(zip(Features, Feature_names, Units))) do (i, (f,n,u))
        plot(df[:,:DateTime],df[:,f];title=n,label=nothing,ylabel=u,size=(800,600),color=i)
    end
    n=length(plots)
    @show n
    plot(plots...;layout=(Int(n/2),Int(n/2)))
end
plot_data(data)
```
